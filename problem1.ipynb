```python
import numpy as np
import matplotlib.pyplot as plt
from copy import copy
```

# Dataset

```python
def dataset4(seed = 42):
    '''
    Dataset IVを返す
    '''
    np.random.seed(seed)
    n = 200
    x = 3 * (np.random.randn(n, 4) - 0.5)
    y = (2 * x[:, 1] - 1* x[:, 2] + 0.5 + 0.5 * np.random.randn(n)) > 0
    y = 2 * y -1
    
    return x.T, y
```

```python
x, y = dataset4()
```

```python
plt.scatter(x[0, :], x[1, :], c=y)
plt.xlabel('x0')
plt.ylabel('x1')
plt.show()
```

# Def functions

```python
la = 0.01
```

```python
def cost_function(x,y,w):
    return np.sum(np.log(1+np.exp(y*np.dot(w,x)))) + la*np.dot(w,w)
```

```python
def gradient(x,y,w):
    return np.sum(-y*x / (np.exp(y*np.dot(w,x))+1),axis=1) + 2*la*w
```

```python
def hessian(x,y,w):
    xxt = np.array([np.dot(xi.reshape(xi.shape[0], 1), xi.reshape(1, xi.shape[0]))for xi in x.T]).transpose(1,2,0)
    u = np.exp(-y*np.dot(w,x))
    
    return np.sum((u/(1+u)**2) * y**2*xxt, axis=2) + 2*la
```

```python
def show_plot(x, y, w=np.array([0])):
    plt.scatter(x[0, :], x[1, :], c=y)
    plt.xlabel('x0')
    plt.ylabel('x1')
    if w.shape[0] == 3:
        x1 = np.linspace(-10, 10,  21)
        x2 = -(x1*w[0] + w[2]) / w[1]
        plt.scatter(x[0, :], x[1, :], c=y)
        plt.plot(x1, x2)
    plt.show()
```

# Problem
## 1. batch steepest gradient method

```python
def batch_steepest_gradient(x, y, w, eta0 = 0.01): 
    eta = copy(eta0)
    min_w = None
    min_J = np.inf
    i = 0
    
    while i < 100:
        value = cost_function(x, y, w)
        if value < min_J:
            min_w = w
            min_J = value
            i = 0
            eta = copy(eta0)
        else:
            i += 1
            eta *= 0.9
            
        w -= eta * gradient(x, y, w) 
        
    return min_w
```

```python
x, y = dataset4()
x = np.array([x[0], x[1], np.ones(x.shape[1])])
w0 = np.random.randn(3)
w = batch_steepest_gradient(x, y, w0)
```

```python
print(w)
```

```python
show_plot(x, y, w)
```

## 2.Newton based method

```python
def newton_based_method(x, y, w, eta0=0.01):
    eta = copy(eta0)
    min_w = None
    min_J = np.inf
    i = 0
    
    while i < 100:
        value = cost_function(x, y, w)
        if value < min_J:
            min_w = w
            min_J = value
            i = 0
            eta = copy(eta0)
        else:
            i += 1
            eta *= 0.9
        w -= eta * np.dot(np.linalg.inv(hessian(x, y, w)), gradient(x, y, w))
        
    return min_w
```

```python
x, y = dataset4()
x = np.array([x[0], x[1], np.ones(x.shape[1])])
w0 = np.random.randn(3)
w = newton_based_method(x, y, w0)
```

```python
print(w)
```

```python
show_plot(x, y, w)
```

## 3. Compare the performance of the above two optimization methods

```python
def compare_batch_steepest_gradient(x, y, w, eta0 = 0.01): 
    eta = copy(eta0)
    min_w = None
    min_J = np.inf
    i = 0
    jts = []
    
    while i < 100:
        value = cost_function(x, y, w)
        if value < min_J:
            min_w = w
            min_J = value
            i = 0
            eta = copy(eta0)
            jts.append(min_J)
        else:
            i += 1
            eta *= 0.9
            
        w -= eta * gradient(x, y, w) 
        
    return min_w, np.array(jts)
```

```python
def compare_newton_based_method(x, y, w, eta0=0.01):
    eta = copy(eta0)
    min_w = None
    min_J = np.inf
    i = 0
    jts = []
    
    while i < 100:
        value = cost_function(x, y, w)
        if value < min_J:
            min_w = w
            min_J = value
            i = 0
            eta = copy(eta0)
            jts.append(min_J)
        else:
            i += 1
            eta *= 0.9
        w -= eta * np.dot(np.linalg.inv(hessian(x, y, w)), gradient(x, y, w))
        
    return min_w, np.array(jts)
```

```python
x, y = dataset4()
x = np.array([x[0], x[1], np.ones(x.shape[1])])
w0 = np.random.randn(3)
w_bsg, jts_bsg = compare_batch_steepest_gradient(x, y, copy(w0), 0.01)
w_new, jts_new = compare_newton_based_method(x, y, w0)
```

```python
print(f'w_bsg: {w_bsg}')
print(f'w_new: {w_new}')
```

```python
def calc_sub_jt(x, y, w, jts):
    sub_jts = []
    for jt in jts:
        value = cost_function(x, y, w)
        sub = abs(jt - value)
        sub_jts.append(sub)
        
    return np.array(sub_jts)
```

```python
sub_jts_bsg = calc_sub_jt(x, y, w_bsg, jts_bsg)
sub_jts_new = calc_sub_jt(x, y, w_new, jts_new)
```

```python
n = min(sub_jts_bsg.shape[0], sub_jts_new.shape[0])
plt.plot(np.arange(n), sub_jts_bsg[:n], label="Batch steepest gradient method")
plt.plot(np.arange(n), sub_jts_new[:n], label="Newton based method")
plt.xlabel('epoch')
plt.ylabel(r'|$J(w^{(t)}-J(\hat{w})}$|')
plt.yscale('log')
plt.legend()
plt.show()
```
